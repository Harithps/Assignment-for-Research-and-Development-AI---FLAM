```python
import numpy as np
import pandas as pd
from scipy.optimize import minimize
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv("xy_data.csv")
x_obs = data["x"].values
y_obs = data["y"].values

# Generate t values linearly between 6 and 60
t = np.linspace(6, 60, len(x_obs))

# Define model
def model(params, t):
    theta_deg, M, X, = params
    theta = np.deg2rad(theta_deg)
    x = t * np.cos(theta) - np.exp(M * np.abs(t)) * np.sin(0.3 * t) * np.sin(theta) + X
    y = 42 + t * np.sin(theta) + np.exp(M * np.abs(t)) * np.sin(0.3 * t) * np.cos(theta)
    return x, y

# Objective: minimize L1 distance
def objective(params):
    x_pred, y_pred = model(params, t)
    return np.sum(np.abs(x_obs - x_pred) + np.abs(y_obs - y_pred))

# Initial guess and bounds
init_guess = [35, 0, 70]
bounds = [(30, 40), (-0.05, 0.05), (0, 100)]


# Optimization
result = minimize(objective, init_guess, bounds=bounds, method='L-BFGS-B')

# Extract optimized parameters
theta_opt, M_opt, X_opt = result.x
print(f"Optimal parameters:\nθ = {theta_opt:.3f}°\nM = {M_opt:.5f}\nX = {X_opt:.3f}")

# Evaluate and visualize fit
x_fit, y_fit = model(result.x, t)
plt.figure(figsize=(7,5))
plt.scatter(x_obs, y_obs, s=10, label="Observed")
plt.plot(x_fit, y_fit, color='red', label="Fitted Model")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Curve Fitting Result")
plt.show()

# Compute mean L1 distance
L1_distance = np.mean(np.abs(x_obs - x_fit) + np.abs(y_obs - y_fit))
print("Mean L1 Distance:", L1_distance)
```
